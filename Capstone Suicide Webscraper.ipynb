```python
def get_posts(x):
    block = x.split()
    
    for i, word in enumerate(block):
        if 'post' in word.lower():
            return block[i + 1]
```

```python
suicide.posts=suicide.posts.apply(get_posts)
```

```python
suicide
```

```python
def get_swedish_fish(x):
    block = x.split()
    
    for i, word in enumerate(block):
        if 'fish' in word.lower():
            try:
                return block[i + 1]
            except: 
                None
```

```python
suicide.swedishfish=suicide.swedishfish.apply(get_swedish_fish)
```
